```python
import pandas as pd
import numpy as np
import os
import plotly.express as px
from scipy.signal import find_peaks
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
df_shehjar_second_sensor1 = pd.read_csv("/Users/shehjarsadhu/Desktop/UniversityOfRhodeIsland/BSN_2024/MedDock/Results_Discussion/Shehjar_second_sensor_data_20240517-140204.csv")
```

```python
df_shehjar_second_sensor1
```

```python
x = df_shehjar_second_sensor1["FSR"]
peaks_pill_drop, _ = find_peaks(x, height=300,distance=220)
```

```python

fig = make_subplots()

# Adding the main line plot
fig.add_trace(go.Scatter(y=x, mode='lines', name='FSR'))

# Adding the peaks
fig.add_trace(go.Scatter(x=peaks_pill_drop, y=x[peaks_pill_drop], mode='markers', marker=dict(symbol='x', size=10), name='Peaks'))

# Adding the zero line
# fig.add_trace(go.Scatter(y=np.zeros_like(x), mode='lines', line=dict(dash='dash', color='gray'), name='Zero Line'))

# Show plot
fig.show()
```
